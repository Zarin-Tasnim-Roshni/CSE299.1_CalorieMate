```python
#For Renaming only 1 Directory
import os

# Define the directory path where your files are located
directory_path = r'C:/Users/Rafin/OneDrive/Desktop/Rice'

# Define a new name pattern for the files
new_name_pattern = '{}.jpg'  # You can adjust the pattern as needed

# Get a list of all the files in the directory
files = os.listdir(directory_path)

# Initialize a counter to generate unique names (if needed)
counter = 100

# Loop through the files and rename them
for file_name in files:
    # Construct the new file name
    new_file_name = new_name_pattern.format(counter)
    
    # Build the full paths for the old and new file names
    old_file_path = os.path.join(directory_path, file_name)
    new_file_path = os.path.join(directory_path, new_file_name)
    
    # Rename the file
    os.rename(old_file_path, new_file_path)
    
    # Increment the counter for the next file (if needed)
    counter += 1

print("All files renamed successfully.")
```

```python
#For Renaming Multiple Directory
import os
from PIL import Image

def rename_files_in_directory(directory_path, new_name_pattern):
    # Get a list of all the files in the current directory
    files = os.listdir(directory_path)

    # Initialize a counter to generate unique names (if needed)
    counter = 1

    # Loop through the files and rename them
    for file_name in files:
        # Construct the new file name
        new_file_name = new_name_pattern.format(counter)

        # Build the full paths for the old and new file names
        old_file_path = os.path.join(directory_path, file_name)
        new_file_path = os.path.join(directory_path, new_file_name)
        
        # Check if the new file name already exists
        if not os.path.exists(new_file_path):
            # Check if the file has a ".png" extension
            if file_name.lower().endswith('.png'):
                # Delete the PNG file
                os.remove(old_file_path)
                print(f"Deleted: {file_name}")
            else:
                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f"Renamed: {file_name} -> {new_file_name}")
        else:
            print(f"Skipped: {file_name} (file with the same name already exists)")

        # Increment the counter for the next file (if needed)
        counter += 1

    print(f"All files in directory '{directory_path}' renamed successfully.")

def rename_files_in_folders(parent_directory, new_name_pattern):
    # Loop through all subdirectories in the parent directory
    for root, dirs, _ in os.walk(parent_directory):
        for folder in dirs:
            folder_path = os.path.join(root, folder)
            rename_files_in_directory(folder_path, new_name_pattern)

# Define the parent directory path where you want to rename files in subfolders
parent_directory_path = r'C:/Users/Rafin/OneDrive/Desktop/Bangla_Food58_V2/valid'

# Define the new name pattern for the files
new_name_pattern = '{}.jpg'  # You can adjust the pattern as needed

# Call the function to rename files in folders within the parent directory
rename_files_in_folders(parent_directory_path, new_name_pattern)
```
